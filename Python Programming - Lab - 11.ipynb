### 01) WAP to demonstrate the use of Pie chart.

```python
fruits = ["Apples", "Oranges", "Bananas", "Grapes"]
quantities = [25, 30, 15, 20]
plt.pie(quantities, labels=fruits,autopct="%1.1f%%",startangle=90)
plt.title("Fruit Quantitie Chart")
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import random

num_points = 100

x_coordinates = [random.uniform(0, 10) for _ in range(num_points)]
y_coordinates = [random.uniform(0, 10) for _ in range(num_points)]


plt.scatter(x_coordinates, y_coordinates)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Random Points")

plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
plt.figure(figsize=(8, 6))
plt.bar(fruits, quantities, color=['red', 'green', 'orange', 'purple'])
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.title("Fruit Quantities")
plt.xticks(rotation=45) 
plt.tight_layout()
```

### 04) WAP to demonstrate the use of Histogram.

```python
import random
data = [random.randint(1, 100) for _ in range(100)]
plt.figure(figsize=(8, 6))
plt.hist(data, bins=10, edgecolor='black') 
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.title("Histogram of Data")
plt.tight_layout()
plt.show()
plt.close()
```

```python

import numpy as np

objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', '.NET')
y_pos = np.arange(len(objects))
performance = [10, 8, 6, 4, 3, 2]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

for bar in plt.gca().patches:
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_y() + bar.get_height() - 1,
             str(bar.get_height()), ha='center')
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import numpy as np
x_data = np.random.rand(100)
y_data = np.random.rand(100)

colors = np.random.rand(100, 3)

plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, c=colors, s=50, edgecolors='black')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Multiple Colors")
plt.tight_layout()


plt.show()
```

### 03) WAP to Display an Image in Grayscale in Matplotlib.

```python
from PIL import Image
import matplotlib.pyplot as plt

def display_grayscale_image(image_path):

  img = Image.open(image_path).convert('L')
  plt.imshow(img, cmap='gray')
  plt.axis('off') 
  plt.show()
image_path = 'akshardham.jpg'
display_grayscale_image(image_path)
```
